# **GRADIENT DESCENT ALGORITHM WITH GRADIENT DIRECTION MOVEMENT AND FOR EACH ITERATION, PARAMETER UPDATION**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Objective function: f(x) = x^2
def f(x):
    return x**2
```

```python
# Derivative of f(x): f'(x) = 2x
def df(x):
    return 2*x
```

```python
# INITIAL PARAMETERS
learning_rate = 0.8   # Step size (η)
x = 4.0               # Initial starting point
iterations = 15       # Number of iterations
```

image.png

```python
#for plotting parabola
x_vals = np.linspace(-5, 5, 500)

#Computing corresponding f(x) values
y_vals = f(x_vals)
```

```python
results = []
```

image.png

```python
# START GRADIENT DESCENT LOOP
plt.figure(figsize=(9,7), dpi=120) #plotting window
```

```python
for i in range(iterations):
    x_current = x
    y_current = f(x_current)
    grad = df(x_current)
    step_size = learning_rate * grad
    x_new = x_current - step_size
    y_new = f(x_new)
    if grad > 0:
        direction = "Left"
    else:
        direction = "Right"
    print(f"\nIteration {i+1}")
    print("-"*60)
    print(f"x = {x_current:.5f}")
    print(f"f(x) = ({x_current:.5f})² = {y_current:.5f}")

    print(f"f'(x) = 2 × {x_current:.5f} = {grad:.5f}")

    print(f"Step Size = η × f'(x) = {learning_rate} × {grad:.5f} = {step_size:.5f}")

    print(f"x_new = x - ηf'(x)")
    print(f"x_new = {x_current:.5f} - {step_size:.5f} = {x_new:.5f}")

    print(f"f(x_new) = ({x_new:.5f})² = {y_new:.5f}")

    print(f"Movement Direction: {direction}")


    # ============================
    # SAVE DATA
    # ============================

    results.append([
        i+1,
        x_current,
        grad,
        x_new,
        y_new,
        abs(step_size),
        direction
    ])


    # ============================
    # CLEAR PREVIOUS PLOT
    # ============================

    plt.clf()    # Remove previous frame


    # ============================
    # PLOT FUNCTION
    # ============================

    plt.plot(
        x_vals,
        y_vals,
        label="f(x) = x²",
        linewidth=2
    )


    # ============================
    # PLOT CURRENT POINT
    # ============================

    plt.scatter(
        x_current,
        y_current,
        color="red",
        s=120,
        label="Current Point"
    )


    # ============================
    # PLOT NEXT POINT
    # ============================

    plt.scatter(
        x_new,
        y_new,
        color="green",
        s=120,
        label="Next Point"
    )


    # ============================
    # DRAW LARGE ARROW
    # ============================

    plt.arrow(
        x_current,
        y_current,
        x_new - x_current,
        y_new - y_current,
        head_width=0.15,           # Big arrow head
        linewidth=3,              # Thick line
        color="black",
        length_includes_head=True
    )


    # ============================
    # SHOW COORDINATES
    # ============================

    plt.text(
        x_current,
        y_current + 0.6,
        f"({x_current:.2f}, {y_current:.2f})",
        fontsize=10
    )

    plt.text(
        x_new,
        y_new + 0.6,
        f"({x_new:.2f}, {y_new:.2f})",
        fontsize=10
    )


    # ============================
    # ADD MOVEMENT DIRECTION
    # ============================

    plt.text(
        -4.5,
        18,
        f"Direction: {direction}",
        fontsize=12,
        color="blue"
    )


    # ============================
    # LABELS AND FORMATTING
    # ============================

    plt.title(f"Gradient Descent - Iteration {i+1}")

    plt.xlabel("x")
    plt.ylabel("f(x)")

    plt.legend()
    plt.grid(True)

    plt.xlim(-5,5)
    plt.ylim(-1,20)


    # ============================
    # PAUSE (2 SECONDS)
    # ============================

    plt.pause(2)


    # ============================
    # UPDATE X
    # ============================

    x = x_new


plt.show()
## TABULAR SUMMARY (AFTER LOOP)
print("\nFINAL SUMMARY TABLE")
print("="*95)

print("Iter |   x       |  f'(x)    |  x_new    |  f(x_new)  | Step Size | Direction")
print("-"*95)

for row in results:
    print(
        f"{row[0]:>4} | "
        f"{row[1]:>8.4f} | "
        f"{row[2]:>8.4f} | "
        f"{row[3]:>8.4f} | "
        f"{row[4]:>9.4f} | "
        f"{row[5]:>9.4f} | "
        f"{row[6]}"
    )
```

# **Gradient Descent Learning Rate Comparison for f(x)=x²**

```python
# =====================================
# IMPORT LIBRARIES
# =====================================

import numpy as np
import matplotlib.pyplot as plt


# =====================================
# FUNCTION AND DERIVATIVE
# =====================================

# Objective function
def f(x):
    return x**2


# Gradient
def df(x):
    return 2*x


# =====================================
# GRADIENT DESCENT FUNCTION
# =====================================

def gradient_descent(lr, x0, iterations):

    x = x0
    history = []   # Store trajectory

    for i in range(iterations):

        fx = f(x)
        grad = df(x)

        x_new = x - lr * grad
        fx_new = f(x_new)

        history.append((x, fx, x_new, fx_new))

        x = x_new

    return history


# =====================================
# PARAMETERS
# =====================================

x0 = 4
iterations = 15


# Learning rates for different cases
lr_optimal = 0.5       # Optimal
lr_small = 0.05        # Too small
lr_high = 0.9          # Overshoot but converge
lr_very_high = 1.5     # Divergence


# Run gradient descent
hist_optimal = gradient_descent(lr_optimal, x0, iterations)
hist_small = gradient_descent(lr_small, x0, iterations)
hist_high = gradient_descent(lr_high, x0, iterations)
hist_diverge = gradient_descent(lr_very_high, x0, iterations)


# =====================================
# PREPARE FUNCTION CURVE
# =====================================

x_vals = np.linspace(-6, 6, 500)
y_vals = f(x_vals)


# =====================================
# PLOTTING FUNCTION
# =====================================

def plot_case(ax, history, title):

    # Plot function
    ax.plot(x_vals, y_vals, linewidth=2)


    # Plot arrows
    for i in range(len(history)):

        x1, y1, x2, y2 = history[i]

        # Current point
        ax.scatter(x1, y1, color="red", s=50)

        # Next point
        ax.scatter(x2, y2, color="green", s=50)

        # Arrow
        ax.arrow(
            x1, y1,
            x2-x1, y2-y1,
            head_width=0.15,
            linewidth=2,
            color="black",
            length_includes_head=True
        )


    # Formatting
    ax.set_title(title, fontsize=11)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(True)
    ax.set_xlim(-6,6)
    ax.set_ylim(-1,40)


# =====================================
# CREATE 2×2 SUBPLOT
# =====================================

fig, axs = plt.subplots(2,2, figsize=(12,10), dpi=130)


# Case 1: Optimal
plot_case(
    axs[0,0],
    hist_optimal,
    "Optimal Learning Rate (η = 0.5)\nFast Convergence"
)


# Case 2: Too Small
plot_case(
    axs[0,1],
    hist_small,
    "Too Small Learning Rate (η = 0.05)\nSlow Convergence"
)


# Case 3: High but Stable
plot_case(
    axs[1,0],
    hist_high,
    "High Learning Rate (η = 0.9)\nOvershoot but Converges"
)


# Case 4: Very Large
plot_case(
    axs[1,1],
    hist_diverge,
    "Very Large Learning Rate (η = 1.5)\nDivergence"
)


# =====================================
# MAIN TITLE
# =====================================

fig.suptitle(
    "Gradient Descent Learning Rate Comparison for f(x)=x²",
    fontsize=16,
    fontweight="bold"
)


plt.tight_layout()
plt.show()
```

# **Gradient Descent Learning Rate Comparison with Step-by-Step Calculation**

```python
# =====================================
# IMPORT LIBRARIES
# =====================================

import numpy as np                     # Numerical computations
import matplotlib.pyplot as plt        # Plotting graphs


# =====================================
# FUNCTION AND DERIVATIVE
# =====================================

# Objective function: f(x) = x²
def f(x):
    return x**2


# Gradient of f(x): f'(x) = 2x
def df(x):
    return 2*x


# =====================================
# GRADIENT DESCENT WITH CALCULATIONS
# =====================================

def gradient_descent(lr, x0, iterations, name):

    x = x0
    history = []       # Stores trajectory for plotting

    print(f"\n===============================")
    print(f"{name}  (η = {lr})")
    print(f"===============================")


    for i in range(iterations):

        # -----------------------------
        # Step 1: Current value
        # -----------------------------

        fx = f(x)                   # f(x) = x²
        grad = df(x)                # f'(x) = 2x


        # -----------------------------
        # Step 2: Step Size
        # -----------------------------

        step_size = lr * grad       # η × f'(x)


        # -----------------------------
        # Step 3: Update Rule
        # -----------------------------

        # x_new = x - ηf'(x)
        x_new = x - step_size

        fx_new = f(x_new)           # f(x_new)


        # -----------------------------
        # Step 4: Direction
        # -----------------------------

        if grad > 0:
            direction = "Left"
        else:
            direction = "Right"


        # =============================
        # PRINT MATH CALCULATION
        # =============================

        print(f"\nIteration {i+1}")
        print("--------------------------------")

        print(f"x = {x:.5f}")

        print(f"f(x) = ({x:.5f})² = {fx:.5f}")

        print(f"f'(x) = 2 × {x:.5f} = {grad:.5f}")

        print(f"Step Size = η × f'(x)")
        print(f"          = {lr} × {grad:.5f}")
        print(f"          = {step_size:.5f}")

        print(f"x_new = x - ηf'(x)")
        print(f"      = {x:.5f} - {step_size:.5f}")
        print(f"      = {x_new:.5f}")

        print(f"f(x_new) = ({x_new:.5f})² = {fx_new:.5f}")

        print(f"Movement Direction: {direction}")


        # -----------------------------
        # Save values
        # -----------------------------

        history.append(
            (x, fx, x_new, fx_new, direction)
        )


        # -----------------------------
        # Update x
        # -----------------------------

        x = x_new


    return history


# =====================================
# PARAMETERS
# =====================================

x0 = 4
iterations = 15


# Learning rates
lr_optimal = 0.5
lr_small = 0.05
lr_high = 0.9
lr_very_high = 1.5


# =====================================
# RUN GRADIENT DESCENT
# =====================================

hist_optimal = gradient_descent(lr_optimal, x0, iterations, "Optimal LR")
hist_small = gradient_descent(lr_small, x0, iterations, "Small LR")
hist_high = gradient_descent(lr_high, x0, iterations, "High LR")
hist_diverge = gradient_descent(lr_very_high, x0, iterations, "Very High LR")


# =====================================
# PREPARE FUNCTION CURVE
# =====================================

x_vals = np.linspace(-6,6,500)
y_vals = f(x_vals)


# =====================================
# PLOTTING FUNCTION
# =====================================

def plot_case(ax, history, title):

    # Plot parabola
    ax.plot(x_vals, y_vals, linewidth=2, label="f(x)=x²")


    # Plot trajectory
    for i in range(len(history)):

        x1, y1, x2, y2, direction = history[i]


        # Current point
        ax.scatter(x1, y1, color="red", s=60)


        # Next point
        ax.scatter(x2, y2, color="green", s=60)


        # Large arrow
        ax.arrow(
            x1, y1,
            x2-x1, y2-y1,
            head_width=0.25,        # Bigger arrow
            linewidth=3,
            color="black",
            length_includes_head=True
        )


    # Legend (dummy points)
    ax.scatter([],[],color="red",label="Current Point")
    ax.scatter([],[],color="green",label="Next Point")
    ax.plot([],[],color="black",label="Update Direction")


    # Formatting
    ax.set_title(title, fontsize=11)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(True)
    ax.legend()

    ax.set_xlim(-6,6)
    ax.set_ylim(-1,40)


# =====================================
# CREATE SUBPLOTS
# =====================================

fig, axs = plt.subplots(2,2, figsize=(13,11), dpi=130)


# Case 1
plot_case(
    axs[0,0],
    hist_optimal,
    "Optimal Learning Rate (η = 0.5)"
)


# Case 2
plot_case(
    axs[0,1],
    hist_small,
    "Too Small Learning Rate (η = 0.05)"
)


# Case 3
plot_case(
    axs[1,0],
    hist_high,
    "High Learning Rate (η = 0.9)"
)


# Case 4
plot_case(
    axs[1,1],
    hist_diverge,
    "Very Large Learning Rate (η = 1.5)"
)


# =====================================
# MAIN TITLE
# =====================================

fig.suptitle(
    "Gradient Descent Learning Rate Comparison with Step-by-Step Calculation",
    fontsize=16,
    fontweight="bold"
)


plt.tight_layout()
plt.show()
```

1.  The learning rate is optimal, and the model converges to the minimum.
2.  The learning rate is too small. It takes more time but converges to the minimum.

3.   The learning rate is higher than the optimal value. It overshoots but converges ( 1/C < η <2/C).
4.   The learning rate is very large. It overshoots and diverges, moves away from the minima, and performance decreases in learning.

# **Loss vs Epoch for Different Learning Rates for function f(x)=x^2**

```python
# Loss comparison
# =====================================
# IMPORT LIBRARIES
# =====================================

import numpy as np
import matplotlib.pyplot as plt


# =====================================
# FUNCTION AND DERIVATIVE
# =====================================

# Loss function
def f(x):
    return x**2


# Derivative
def df(x):
    return 2*x


# =====================================
# GRADIENT DESCENT WITH LOSS TRACKING
# =====================================

def gradient_descent_loss(lr, x0, epochs, name):

    x = x0
    losses = []

    print(f"\n==============================")
    print(f"{name} (η = {lr})")
    print("==============================")


    for epoch in range(epochs):

        # ----------------------------
        # Step 1: Current values
        # ----------------------------

        fx = f(x)
        grad = df(x)


        # ----------------------------
        # Step 2: Step size
        # ----------------------------

        step_size = lr * grad


        # ----------------------------
        # Step 3: Update
        # ----------------------------

        x_new = x - step_size

        fx_new = f(x_new)


        # ----------------------------
        # Save loss
        # ----------------------------

        losses.append(fx_new)


        # ============================
        # PRINT CALCULATION
        # ============================

        print(f"\nEpoch {epoch+1}")
        print("---------------------------")

        print(f"x = {x:.5f}")
        print(f"f(x) = ({x:.5f})² = {fx:.5f}")

        print(f"f'(x) = 2×{x:.5f} = {grad:.5f}")

        print(f"Step Size = {lr} × {grad:.5f} = {step_size:.5f}")

        print(f"x_new = {x:.5f} - {step_size:.5f} = {x_new:.5f}")

        print(f"Loss = f(x_new) = ({x_new:.5f})² = {fx_new:.5f}")


        # ----------------------------
        # Update x
        # ----------------------------

        x = x_new


    return losses


# =====================================
# PARAMETERS
# =====================================

x0 = 4
epochs = 25


# Learning rates
lr_good = 0.5        # Optimal
lr_low = 0.05        # Too small
lr_high = 0.9        # High
lr_very_high = 1.5   # Divergent


# =====================================
# RUN TRAINING
# =====================================

loss_good = gradient_descent_loss(lr_good, x0, epochs, "Good Learning Rate")

loss_low = gradient_descent_loss(lr_low, x0, epochs, "Low Learning Rate")

loss_high = gradient_descent_loss(lr_high, x0, epochs, "High Learning Rate")

loss_very_high = gradient_descent_loss(lr_very_high, x0, epochs, "Very High Learning Rate")


# =====================================
# PLOT LOSS COMPARISON
# =====================================

plt.figure(figsize=(9,7), dpi=120)


epochs_list = range(1, epochs+1)


# Plot curves
plt.plot(epochs_list, loss_good, label="Good Learning Rate (η=0.5)", linewidth=2)

plt.plot(epochs_list, loss_low, label="Low Learning Rate (η=0.05)", linewidth=2)

plt.plot(epochs_list, loss_high, label="High Learning Rate (η=0.9)", linewidth=2)

plt.plot(epochs_list, loss_very_high, label="Very High Learning Rate (η=1.5)", linewidth=2)


# Labels
plt.xlabel("Epoch")
plt.ylabel("Loss  f(x)")
plt.title("Loss vs Epoch for Different Learning Rates")


# Grid and legend
plt.grid(True)
plt.legend()


plt.show()
```

# **General Plot Showing Effect of Learning Rate on Training Loss **

```python
import numpy as np
import matplotlib.pyplot as plt


# ============================
# EPOCHS
# ============================

epochs = 100
x = np.arange(epochs)


# ============================
# SYNTHETIC LOSS CURVES
# (Made to resemble textbook image)
# ============================

# Good learning rate (fast smooth decay)
good = np.exp(-0.08 * x)

# Low learning rate (slow decay)
low = np.exp(-0.02 * x)

# High learning rate (oscillating decay)
high = np.exp(-0.05 * x) * (1 + 0.2 * np.sin(0.3 * x))

# Very high learning rate (diverges)
very_high = np.exp(0.05 * x) / 50


# ============================
# PLOT
# ============================

plt.figure(figsize=(8,6), dpi=130)


# Plot curves
plt.plot(good, linewidth=2, label="Good learning rate")
plt.plot(low, linewidth=2, label="Low learning rate")
plt.plot(high, linewidth=2, label="High learning rate")
plt.plot(very_high, linewidth=2, label="Very high learning rate")


# Labels
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.title("Effect of Learning Rate on Training Loss")


# Remove box (like reference image)
plt.box(False)


# Grid off (reference has no grid)
plt.grid(False)


# Legend
plt.legend(loc="upper left")


plt.show()
```
